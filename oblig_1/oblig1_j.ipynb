```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Oppgave 2

x = np.linspace(0,40,100)

f = x**(1/2)
f0 = np.ones(100); f0 = f0*2
f1 = 2 + (1/4)*(x-4)
f2 = 2 + (1/4)*(x-4) - (1/64) * (x-4)**2

plt.plot(x, f, color="lightblue", label="f(x)")
plt.plot(x, f0, color="blue", label="T0(x)")
plt.plot(x, f1, color="purple", label="T1(x)")
plt.plot(x, f2, color="pink", label="T2(x)")
plt.plot(4,2, color="black", marker="o", label="(4, 2)")

plt.grid()
plt.legend()
plt.show()
```

```python
#Oppgave 3
#Plotting the values and the regression line

years = np.arange(0,6,1)
xx = np.linspace(0,7,50)
saldo = [5000, 5275, 5592, 5910, 6206, 6553]
plt.plot(years, saldo, color="black", marker="o", label="Saldo")

y = np.matrix(saldo).T
A = np.matrix([ [1, 0],
                [1, 1],
                [1, 2],
                [1, 3],
                [1, 4],
                [1, 5]
              ])
x = np.matmul( np.matmul( np.linalg.inv(np.matmul(A.T, A)), (A.T) ), (y) )
regression = x[1] * xx + x[0]; regression = regression.T
plt.plot(xx, regression, label="Regression of saldo")

plt.grid()
plt.legend()
plt.show()
```

```python
# Oppgave 3
# Using the logarithm for the estimate ( K(t) = K0 + e^(kt) )
# ln(K) = ln( K0 * e^(kt) ) = ln(K0) + kt

plt.plot(years, np.log(saldo), color="black", marker="o", label="ln(saldo)")

y1 = np.matrix(np.log(saldo)).T
x1 = np.matmul( np.matmul( np.linalg.inv(np.matmul(A.T, A)), (A.T) ), (y1) )
print(x1)
regression1 = (x1[1] * xx + x1[0]).T
plt.plot(xx, regression1, label="Regression of ln(saldo)")

plt.grid()
plt.legend()
plt.show()
```

```python
# Plot of the final exponential function against the original datapoints
K0 = np.exp(x1[0])
k = x1[1]
expfunc = K0 * np.exp(k*xx)

plt.plot(years, saldo, color="black", marker="o", label="Saldo")
plt.plot(xx, expfunc.T, color="lightblue", label="Exponential regression")

plt.grid()
plt.legend()
plt.show()
```
